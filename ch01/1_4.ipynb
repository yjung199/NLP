## 신경망 구현

```python
import sys
sys.path.append('..')  # 부모 디렉터리의 파일을 가져올 수 있도록 설정
import numpy as np
from common.layers import Affine, Sigmoid, SoftmaxWithLoss


class TwoLayerNet:
    # input_size = x의 특성수, in this case "2"
    # hidden_size = 10으로 설정 -> bias의 크기 (10,)
    # (input_size,hidden_size) = (2, 10) = 첫번째 히든 레이어의 크기
    # output_size = 3 -> 다중분류, 3개로 나눠서 표출
    def __init__(self, input_size, hidden_size, output_size):
        I, H, O = input_size, hidden_size, output_size

        # 가중치와 편향 초기화
        W1 = 0.01 * np.random.randn(I, H) # (2,10) 
        b1 = np.zeros(H)                  # (10,), 0 으로 초기화
        W2 = 0.01 * np.random.randn(H, O) # (10,3), 출력층
        b2 = np.zeros(O)                  # (3,)

        # 계층 생성
        self.layers = [
            Affine(W1, b1), # 은닉층
            Sigmoid(),      # 은닉층 activation
            Affine(W2, b2)  # 출력층
        ]
        self.loss_layer = SoftmaxWithLoss() # 출력층에서 출력을 확률로 변환하고 손실까지 계산

        # 모든 가중치와 기울기를 리스트에 모은다.
        self.params, self.grads = [], []
        for layer in self.layers:
            self.params += layer.params # 레이어를 돌면서 각 레이어들의 파라미터 담음 (W, b)
            self.grads += layer.grads   # 레이어를 돌면서 각 레이어들의 미분을 담음 (W의 미분, b의 미분)

    def predict(self, x):
        for layer in self.layers:
            x = layer.forward(x)
        return x

    def forward(self, x, t): #정방향으로 자기가 물고 있는 모든 레이어를 따라가서 그 속에 있는 forward() 호출
        score = self.predict(x)
        loss = self.loss_layer.forward(score, t)
        return loss

    def backward(self, dout=1): #dout =1 becasue 자기자신 미분하면 1이므로 제일 끝의 미분은 1. Affine - Sigmoid - Affine - loss_layer 순으로 정방향했기 때문에 똑같이 loss_layer층의 미분 (backward())타기. 
        dout = self.loss_layer.backward(dout)
        for layer in reversed(self.layers):
            dout = layer.backward(dout)
        return dout
```

## 학습용 코드

```python
import sys
sys.path.append('..')  # 부모 디렉터리의 파일을 가져올 수 있도록 설정
import numpy as np
from common.optimizer import SGD
from dataset import spiral
import matplotlib.pyplot as plt
from two_layer_net import TwoLayerNet


# 하이퍼파라미터 설정
max_epoch = 300
batch_size = 30 # 30개 씩 학습
hidden_size = 10
learning_rate = 1.0

# 데이터 읽기, 모델과 옵티마이저 생성
x, t = spiral.load_data()
model = TwoLayerNet(input_size=2, hidden_size=hidden_size, output_size=3)
optimizer = SGD(lr=learning_rate)

# 학습에 사용하는 변수
data_size = len(x)
max_iters = data_size // batch_size # 10번 학습
total_loss = 0
loss_count = 0
loss_list = []

#전체 에포크 (300번)
for epoch in range(max_epoch):
    # 데이터 뒤섞기
    idx = np.random.permutation(data_size)
    x = x[idx]
    t = t[idx]

    # 미니배치
    for iters in range(max_iters):
        batch_x = x[iters*batch_size:(iters+1)*batch_size]
        batch_t = t[iters*batch_size:(iters+1)*batch_size]

        # 기울기를 구해 매개변수 갱신
        loss = model.forward(batch_x, batch_t) # (30,2) 짜리 데이터가 들어감. model.forward 는 전체 모델 다 포함
        model.backward()
        optimizer.update(model.params, model.grads) # 학습이 진행되는 곳 (weight -= weight - a*미분)

        total_loss += loss
        loss_count += 1

        # 정기적으로 학습 경과 출력
        if (iters+1) % 10 == 0:
            avg_loss = total_loss / loss_count
            print('| 에폭 %d |  반복 %d / %d | 손실 %.2f'
                  % (epoch + 1, iters + 1, max_iters, avg_loss))
            loss_list.append(avg_loss)
            total_loss, loss_count = 0, 0

# plt.rc('font', family='Malgun Gothic')

# 학습 결과 플롯
plt.plot(np.arange(len(loss_list)), loss_list, label='train')
plt.xlabel('반복 (x10)')
plt.ylabel('손실')
plt.show()

# 경계 영역 플롯
h = 0.001
x_min, x_max = x[:, 0].min() - .1, x[:, 0].max() + .1
y_min, y_max = x[:, 1].min() - .1, x[:, 1].max() + .1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
X = np.c_[xx.ravel(), yy.ravel()]
score = model.predict(X)
predict_cls = np.argmax(score, axis=1)
Z = predict_cls.reshape(xx.shape)
plt.contourf(xx, yy, Z)
plt.axis('off')

# 데이터점 플롯
x, t = spiral.load_data()
N = 100
CLS_NUM =    3
markers = ['o', 'x', '^']
for i in range(CLS_NUM):
    plt.scatter(x[i*N:(i+1)*N, 0], x[i*N:(i+1)*N, 1], s=40, marker=markers[i])
plt.show()
```
